```python
import math
def Gaussian_cdf(data):
    sorted_data=np.sort(data)
    CDF_values=[]
    mean=np.mean(sorted_data)
    sd=np.std(sorted_data)
    for i in sorted_data:
        k=(i-mean)/sd*np.sqrt(2)
        erf_value=math.erf(k)
        final_result=1/2*(1+erf_value)
        CDF_values.append(final_result)
    return CDF_values
```

```python
data=df["SepalLengthCm"]
cdf_values=Gaussian_cdf(data)
```

```python
cdf_array=np.array(cdf_values)
```

```python
sorted_x=np.sort(data)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.xlabel("X_values(X)")
plt.ylabel("CDF_values(F(X)")
plt.title("Gaussian CDF")
plt.plot(sorted_x,cdf_array)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
data = np.array([145,135,157,156,157,139,140,156,155,110])
# sort data
sorted_data = np.sort(data)
# cumulative probabilities
cdf = np.arange(1, len(sorted_data)+1) / len(sorted_data)
print("Sorted Data:", sorted_data)
print("CDF:", cdf)
plt.plot(sorted_data, cdf, marker='o')
plt.xlabel("Data Values")
plt.ylabel("CDF")
plt.title("Cumulative Distribution Function")
plt.show()
```
