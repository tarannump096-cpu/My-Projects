```python
def stimulate_CLT(data,no_of_samples,sample_size):
    samples_mean=[]
    for i in range(no_of_samples):
        samples=np.random.choice(data,size=sample_size)
        samples_mean.append(np.mean(samples))
    means_array=np.array(samples_mean)
    return np.mean(means_array),means_array
```

```python
data=df["SepalLengthCm"]
no_of_samples=1000
sample_size=30
sample_mean,mean_array=stimulate_CLT(data,no_of_samples,sample_size)
```
