```python
def estimate_CLT_Age(data,no_of_samples,sample_size):
    sample_mean=[]
    for i in range(no_of_samples):
        samples=np.random.choice(data,size=sample_size)
        sample_mean.append(np.mean(samples))
    sample_mean_array=np.array(sample_mean)
    return np.mean(sample_mean_array)
```

```python
import math
def Gaussian_CDF_height(data):
    sorted_data=np.sort(data)
    GCDF_values=[]
    mean=np.mean(sorted_data)
    sd=np.std(sorted_data)
    for i in sorted_data:
        k=(i-mean)/(np.sqrt(2)*sd)
        erf=math.erf(k)
        final=1/2*(1+erf)
        GCDF_values.append(final)
    GCDF_Array=np.array(GCDF_values)
    return GCDF_Array
```

```python
data=df["Height_cm"]
cdf_values=Gaussian_CDF_height(data)
sorted_data=np.sort(data)
```

```python
plt.xlabel("X_values(X)")
plt.ylabel("cdf_values(F(X))")
plt.title("Gaussian CDF for Height")
plt.plot(sorted_data,cdf_values)
plt.show()
```

```python
import pandas as pd
import numpy as np

def empirical_bootstrapping_height(data, no_of_samples, sample_size, CI):
    sample_means = []
    for _ in range(no_of_samples):
        # Generate a resample with replacement
        resample = np.random.choice(data, size=sample_size, replace=True)
        sample_means.append(np.mean(resample))

    # Calculate the confidence interval from the distribution of sample means
    alpha = 1 - CI
    lower_bound = np.percentile(sample_means, (alpha / 2) * 100)
    upper_bound = np.percentile(sample_means, (1 - alpha / 2) * 100)

    return lower_bound, upper_bound, np.array(sample_means)

# Define data here to ensure it's always available for this cell
try:
    df = pd.read_csv("/content/diet_recommendation_dataset_1000.csv") # Added for robust execution
    data = df["Height_cm"]
    num_bootstrap_samples = 1000 # Number of bootstrap samples
    bootstrap_sample_size = 50   # Size of each bootstrap sample
    confidence_level = 0.95

    # Call the function and get the confidence interval
    lb, ub, means = empirical_bootstrapping_height(
        data=data,
        no_of_samples=num_bootstrap_samples,
        sample_size=bootstrap_sample_size,
        CI=confidence_level
    )

    print("{}% Confidence Interval for Height: ({:.2f}, {:.2f})".format(confidence_level*100, lb, ub))

except FileNotFoundError:
    print("Error: The file 'diet_recommendation_dataset_1000.csv' was not found in the /content/ directory.")
    print("Please upload the dataset file to your Colab environment.")
    print("You can upload it by clicking the folder icon on the left sidebar, then the upload icon.")
    print("Alternatively, ensure cell '7sGGmwHkrds4' is executed successfully after uploading the file.")
```
