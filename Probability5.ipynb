```python
import numpy as np
```

```python
alpha=2.0
```

```python
data=np.random.pareto(alpha,size=1000)
```

```python
data
```

```python
import scipy.stats as stats
```

```python
import matplotlib.pyplot as plt
stats.probplot(data,dist=stats.pareto,sparams=(alpha,),plot=plt)
plt.show()
```

```python
from scipy.stats import boxcox
```

```python
transformed_data,lambda_value=stats.boxcox(data)
```

```python
lambda_value
```

```python
transformed_data
```

```python
stats.probplot(transformed_data,dist="norm",plot=plt)
plt.show()
```

```python
import seaborn as sns
```

```python
sns.distplot(transformed_data)
```

```python
from scipy.stats import skew,kurtosis
```

```python
skew=skew(transformed_data)
```

```python
skew
```

```python
kur=kurtosis(transformed_data)
```
