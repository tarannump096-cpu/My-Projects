```python
import numpy as np
import math
def GCDF(data):
    sorted_data=np.sort(data)
    mean=np.mean(data)
    std=np.std(data)
    GCDF_values=[]
    for i in sorted_data:
        k=(i-mean)/(np.sqrt(2)*std)
        erf=math.erf(k)
        final=1/2*(1+erf)
        GCDF_values.append(final)
    GCDF_array=np.array(GCDF_values)
    return GCDF_array
def ECDF(data):
    sorted_data=sorted(data)
    ECDF_values=[]
    n=len(sorted_data)
    i=0
    while i<n:
        var=(i+1)/n
        ECDF_values.append(var)
        i+=1
    ECDF_array=np.array(ECDF_values)
    return ECDF_array
def kstest(data):
    gcdf_values=GCDF(data)
    ecdf_values=ECDF(data)
    n=len(data)
    i=0
    diff=0
    while i<n:
        value1=gcdf_values[i]
        value2=ecdf_values[i]
        D=abs(value1-value2)
        if D>diff:
            diff=D
        i+1
    critical_value=1.385*np.sqrt((n+n)/(n*n))
    return diff,critical_value
```

```python
data=np.random.normal(0,1,1000)
test_stats,critical=kstest(data)
```
