```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("/content/train.csv")
```

```python
df
```

```python
df1=df[["Survived","Pclass"]].dropna()
```

```python
df1
```

```python
first_class_survived=len(df1[(df1["Survived"]==1 )& (df1["Pclass"]==1)])
second_class_survived=len(df1[(df1["Survived"]==1 )& (df1["Pclass"]==2)])
third_class_survived=len(df1[(df1["Survived"]==1 )& (df1["Pclass"]==3)])
first_class_not_survived=len(df1[(df1["Survived"]==0 )& (df1["Pclass"]==1)])
second_class__not_survived=len(df1[(df1["Survived"]==0 )& (df1["Pclass"]==2)])
third_class_not_survived=len(df1[(df1["Survived"]==0)& (df1["Pclass"]==3)])
```

```python
observed_matrix=np.array([[first_class_survived,second_class_survived,third_class_survived],
                          [first_class_not_survived,second_class__not_survived,third_class_not_survived]])
```

```python
observed_matrix
```

```python
from scipy.stats import chi2_contingency
```

```python
chi2,p,dof,expected=chi2_contingency(observed_matrix)
```

```python
p
```

```python
if p>5:
    print("there is no relation")
else:
    print("there is a relation")
```
