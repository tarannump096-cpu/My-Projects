```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Define your sample data
data = [145, 135, 157, 156, 157, 139, 140, 156, 155,110]

# 2. Sort the data in ascending order
x = np.sort(data)

# 3. Calculate cumulative probability (y-axis)
# Each point i represents the fraction (i/n) of data points <= x[i]
n = len(x)
y = np.arange(1, n + 1) / n

# 4. Create the Step Function plot
plt.figure(figsize=(8, 5))
plt.step(x, y, where='post', color='darkblue', linewidth=2, label='ECDF Step')
plt.scatter(x, y, color='green', s=30, label='Data Points', zorder=3)

# Formatting
plt.title('Discrete CDF: Sorted Data vs Cumulative Probability')
plt.xlabel('Sorted Values')
plt.ylabel('P(X ≤ x)')
plt.grid(axis='both', linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

```python
import numpy as np

# 1. Your sample data
data = np.array([12, 15, 14, 18, 20, 22, 19, 17, 16, 21, 25, 23, 14, 18, 19])

def bootstrap_ci(data, n_iterations=10000, ci=95):
    boot_stats = []
    n_size = len(data)

    # 2. Resample with replacement n_iterations times
    for _ in range(n_iterations):
        sample = np.random.choice(data, size=n_size, replace=True)
        boot_stats.append(np.mean(sample)) # Calculate statistic of interest

    # 3. Find the lower and upper percentiles (2.5% and 97.5% for a 95% CI)
    lower_p = (100 - ci) / 2
    upper_p = 100 - lower_p

    lower_bound = np.percentile(boot_stats, lower_p)
    upper_bound = np.percentile(boot_stats, upper_p)

    return lower_bound, upper_bound

# Execution
lower, upper = bootstrap_ci(data)
print(f"95% Confidence Interval: [{lower:.2f}, {upper:.2f}]")
```
