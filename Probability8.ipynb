```python
import numpy as np
def permutation_test(sample1,sample2,no_of_resamples):
    sample1_Array=np.array(sample1)
    sample2_Array=np.array(sample2)
    D=abs(np.mean(sample1_Array)-np.mean(sample2_Array))
    combined_sample=np.concatenate((sample1_Array,sample2_Array))
    n1=len(sample1_Array)
    n2=len(combined_sample)
    diff_lst=[]
    for i in range(no_of_resamples):
        jumbled_sample=np.random.permutation(combined_sample)
        new_sample1=jumbled_sample[:n1]
        new_sample2=jumbled_sample[n1:n2]
        diff=abs(np.mean(new_sample1)-np.mean(new_sample2))
        diff_lst.append(diff)
    diff_array=np.array(diff_lst)
    diff_array_sorted=np.sort(diff_array)
    count=0
    for i in diff_array_sorted:
        if i<D:
            count+=1
    percentage=(count/len(diff_array_sorted))*100
    return percentage
```

```python
sample1=np.random.randint(140,192,50)
sample2=np.random.randint(140,192,50)
no_of_resamples=1000
p_value=permutation_test(sample1,sample2,no_of_resamples)
```
