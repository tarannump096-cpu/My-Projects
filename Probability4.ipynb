```python
import numpy as np
```

```python
def estimate_chebyshevs_non_Gaussian(meu,sd,k):
    LL=meu-k*sd
    UL=meu+k*sd
    percentage=(1-(1/k**2))*100
    return (LL,UL),percentage
```

```python
meu=168
sd=5
k=2  #2nd range of SD for Non Gaussian distribution.
estimate_chebyshevs_non_Gaussian(meu,sd,k)
```
