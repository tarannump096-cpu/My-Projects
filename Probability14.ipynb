```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("diet_recommendation_dataset_1000.csv")
```

```python
df
```

```python
x=df.drop(columns=["Diet"])
```

```python
y=df["Diet"]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=365,stratify=y)
```

```python
ss=StandardScaler()
```

```python
le=LabelEncoder()
```

```python
xtrain_scaled=ss.fit_transform(xtrain)
```

```python
xtest_scaled=ss.transform(xtest)
```

```python
ytrain_scaled=le.fit_transform(ytrain)
```

```python
ytest_scaled=le.transform(ytest)
```

```python
xtrain_scaled.shape
```

```python
xtest_scaled.shape
```

```python
ytrain_scaled.shape
```

```python
from sklearn.feature_selection import SequentialFeatureSelector
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression(max_iter=1000)
```

```python
ffs=SequentialFeatureSelector(
    model,
    n_features_to_select=5,
    direction='forward',
    scoring="accuracy",
    cv=5
)
```

```python
ffs.fit(xtrain_scaled,ytrain_scaled)
```

```python
ffs.get_support()
```

```python
x.columns
```

```python
ffs_features=x.columns[ffs.get_support()]
```

```python
ffs_features
```

```python
bfs=SequentialFeatureSelector(
    model,
    n_features_to_select=5,
    direction='backward',
    scoring="accuracy",
    cv=5
)
```

```python
bfs.fit(xtrain_scaled,ytrain_scaled)
```

```python
bfs_features=x.columns[bfs.get_support()]
```

```python
bfs_features
```
