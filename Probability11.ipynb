```python
import numpy as np
import math
def GCDF(data):
    sorted_data=np.sort(data)
    mean=np.mean(sorted_data)
    sd=np.std(sorted_data)
    cdf_lst=[]
    for i in sorted_data:
        k=(i-mean)/np.sqrt(2)*sd
        erf=math.erf(k)
        final=1/2*(1+erf)
        cdf_lst.append(final)
    cdf_array=np.array(cdf_lst)
    return cdf_array
def ECDF(data):
    sorted_data=sorted(data)
    ecdf_lst=[]
    n=len(data)
    i=0
    while i<n:
        var=(i+1)/n
        ecdf_lst.append(var)
        i+=1
    ecdf_array=np.array(ecdf_lst)
    return ecdf_array
def ks_test(data):
    gcdf_values=GCDF(data)
    ecdf_values=ECDF(data)
    n=len(data)
    D=0
    i=0
    while i<n:
        value1=gcdf_values[i]
        value2=ecdf_values[i]
        diff=abs(value1-value2)
        if diff>D:
            D=diff
        i+=1
    critical_value=1.358*np.sqrt((n+n)/n*n)
    return D,critical_value
```

```python
data=np.random.normal(0,1,100)
d,critical=ks_test(data)
```
