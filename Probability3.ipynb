```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
df=pd.read_csv("laptop_data.csv")
```

```python
sns.distplot(df["Price"])
```

```python
log_data=np.log(df["Price"])
```

```python
stats.probplot(log_data,dist="norm")
```
